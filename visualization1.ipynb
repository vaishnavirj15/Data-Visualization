```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Penguins dataset
penguins = sns.load_dataset('penguins')

# Drop rows with missing values
penguins = penguins.dropna()

# 1. Pair Plot
sns.pairplot(penguins, hue='species', diag_kind='kde', palette='Set2')
plt.suptitle('Pairwise Relationships in Penguins Dataset', y=1.02)
plt.show()

# 2. Boxplot: Distribution of Flipper Length by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=penguins, palette='cool', dodge=False, hue=None)
plt.title('Boxplot: Flipper Length across Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()

# 3. Bar Plot: Average Body Mass by Island
plt.figure(figsize=(8, 6))
sns.barplot(x='island', y='body_mass_g', hue='species', data=penguins, errorbar=None, palette='viridis')
plt.title('Barplot: Body Mass by Island and Species')
plt.xlabel('Island')
plt.ylabel('Body Mass (g)')
plt.legend(title='Species')
plt.show()

# 4. Scatter Plot: Bill Length vs Bill Depth
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='bill_length_mm', y='bill_depth_mm',
    hue='species', style='sex', data=penguins, palette='Dark2', s=100
)
plt.title('Scatter Plot: Bill Length vs Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.grid(alpha=0.5)
plt.show()

# 5. Heatmap: Correlation Matrix
plt.figure(figsize=(8, 6))
correlation_matrix = penguins.drop(columns=['species', 'island', 'sex']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap: Feature Correlations in Penguins Dataset')
plt.show()

# 6. Count Plot: Distribution of Species by Island
plt.figure(figsize=(8, 6))
sns.countplot(x='island', hue='species', data=penguins, palette='Set1')
plt.title('Count Plot: Species Distribution by Island')
plt.xlabel('Island')
plt.ylabel('Count')
plt.legend(title='Species')
plt.show()

```
